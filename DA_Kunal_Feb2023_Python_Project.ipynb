## TASK8:

```python
#appending all the list from porduction_company column and flattening them.
import ast
a1=original_data_backup["production_companies"]
list_comp=[]
for i in range(0,len(original_data_backup)):
    if i in original_data_backup.index:
        a2=original_data_backup["production_companies"][i]
        b2=ast.literal_eval(a2)
        list_comp.append(b2)
list_comp=sum(list_comp,[])
list_comp
```

```python
#removing duplicates from the list
production_companies=list(set(list_comp))
production_companies
```

```python
#Creating dictionary with production company name and no of movies it produced
produc_dict={}
for i in production_companies:
    count = 0
    for j in range(0,len(original_data_backup)):
        if j in original_data_backup.index:
            row_list=original_data_backup["production_companies"][j]
            if i in row_list:
                count+=1
    to_update={i:count}
    produc_dict.update(to_update)
```

```python
#converting dictionary to dataframe
produc_df=pd.DataFrame(produc_dict.values(),index=produc_dict.keys())
produc_df.columns=["No of movie produced"]
produc_df
```

Production Company as index and "no of movies they produced" as values.

## TASK 10:

```python
#Sorting in descresing order based on revenue
filter_500_based_on_rev=original_data_backup.sort_values("revenue",ascending=False).head(500)
filter_500_based_on_rev
```

```python
def central_tendency_of_column(column):
    col_mean=filter_500_based_on_rev[column].mean()
    col_median=filter_500_based_on_rev[column].median()
    col_mode=filter_500_based_on_rev[column].mode()[0]
    return col_mean,col_median,col_mode

col_for_central_tend=["budget","revenue","runtime"]

central_tend_values=map(central_tendency_of_column,col_for_central_tend)
central_df=pd.DataFrame(central_tend_values,columns=["Mean","Median","Mode"],index=["budget","revenue","runtime"])
central_df
```

```python
fig,ax=plt.subplots(1,3,figsize=(16,9))
sns.boxplot(x=filter_500_based_on_rev["budget"],data=filter_500_based_on_rev,ax=ax[0])

sns.boxplot(x=filter_500_based_on_rev["revenue"],data=filter_500_based_on_rev,ax=ax[1])

sns.boxplot(x=filter_500_based_on_rev["runtime"],data=filter_500_based_on_rev,ax=ax[2])

plt.show()
```
